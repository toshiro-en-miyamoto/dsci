```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("data/air_quality_no2.csv", index_col=0, parse_dates=True)
```

```python
air_quality
```

According to [pandas API Reference](https://pandas.pydata.org/docs/reference/io.html),

- `index_col`: Column(s) to use as the row labels of the `DataFrame`, either given as string name or column index. If a sequence of int / str is given, a MultiIndex is used.
- `parse_dates`: if Tue, try parsing the index

With a `DataFrame`, pandas creates by default one line plot for each of the columns with numeric data.

```python
air_quality.plot()
```

To plot a specific column, use the selection method in combination with the `plot()` method. Hence, the `plot()` method works on both `Series` and `DataFrame`.

```python
air_quality["station_paris"].plot()
```

[`DataFrame.plot`](https://pandas.pydata.org/docs/reference/frame.html#plotting) is both a callable method, `plot()`, and a namespace attribute for specific plotting methods of the form `DataFrame.plot.<kind>()`. The kind of plot to produce:

- `line`: line plot (default)
- `bar`: vertical bar plot
- `barh`: horizontal bar plot
- `hist`: histogram
- `box`: boxplot
- `kde`: Kernel Density Estimation plot
- `density`: same as `kde`
- `area`: area plot
- `pie`: pie plot
- `scatter`: scatter plot (DataFrame only)
- `hexbin`: hexbin plot (DataFrame only)

The [Chart Visualization](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html) section of the pandas User Guide provides more details.

```python
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
```

Arguments of the `scatter()` method in the above example;

- `x` and `y`: The column name (*str*) or column position  (*int*) to be used as horizontal and vertical coordinates for each point.
- `alpha`: 0.0 transparent through 1.0 opaque; one of the options to pass to `matplotlib` plotting method, also known as *keyword arguments* or *kwargs*.

Separate subplots for each of the data columns are supported by the `subplots` argument of the `plot` functions.

```python
air_quality.plot.area(figsize=(12, 4), subplots=True)
```

According to the API Reference [`pandas.DataFrame.plot`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html#pandas.DataFrame.plot), the `plot()` method and `plot.<kind>()` methods return an instance of `matplotlib.axes.Axes` or `numpy.ndarray` of them.

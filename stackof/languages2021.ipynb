```python
language = pd.read_csv('language2021.csv')
```

```python
popular_raw = language[language["kind"]=="popular"]
```

```python
popular_pivot = popular_raw.pivot(index="language", columns="year", values="rating")
```

```python
recent_popular = popular_pivot[popular_pivot[2021].notna() & popular_pivot[2020].notna()]
```

```python
recent_popular.sort_values(by=2021,ascending=False)
```
